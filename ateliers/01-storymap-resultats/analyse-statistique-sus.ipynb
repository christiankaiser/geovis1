# Comparaison SUS pour les volées 2017 et 2018

Nous essayons de déterminer ici si le System Usability Score (SUS) pour l'utilisation de l'application [wevis.ch](http://wevis.ch) a changé entre la volée de Géovis 1 de 2017 et de 2018.

Pour ce faire, nous faisons un test statistique de comparaison de moyenne entre 2017 et 2018. Si nos deux échantillons suivent une loi normale, nous pouvons utiliser le test de Student.

Voici les données pour 2017:

```r
sus17 = c(72.5, 72.5, 70, 85, 65, 70, 62.5, 57.5, 55, 77.5, 87.5, 82.5, 72.5, 82.5, 87.5, 87.5, 77.5, 72.5, 90, 77.5, 72.5, 90, 87.5, 67.5, 65, 77.5, 50, 77.5, 77.5, 72.5, 57.5, 92.5, 65, 75, 72.5, 60, 87.5, 47.5, 57.5, 72.5, 80, 90, 77.5, 57.5, 72.5)
```

et pour 2018:

```r
sus18 = c(60, 62.5, 80, 82.5, 65, 60, 47.5, 92.5, 90, 70, 50, 80, 77.5, 67.5, 60, 90, 70, 77.5, 75, 67.5)
```

Pour avoir une idée de la distribution, nous faisons rapidement un histogramme pour chaque année:

```r
hist(sus17)
```

```r
hist(sus18)
```

La distribution pour 2017 semble être moins régulière que 2018. Le nombre de personnes ayant rempli le SUS est différent aussi:

```r
length(sus17)
```

```r
length(sus18)
```

Beaucoup plus de participants en 2017: 45 contre seulement 20 en 2018.

Calculons le test de Kolmogorov-Smirnov pour déterminer si les distributions suivent une loi normale.

Les hypothèses sont comme suit:

$H_0$ : la population de laquelle a été tiré l'échantillon à tester suit la distribution de comparaison (i.e. une loi normale)

$H_1$: la population de laquelle a été tiré l'échantillon ne suit pas la distribution de comparaison

Ceci donne pour 2017:

```r
ks.test(sus17, "pnorm", mean=mean(sus17), sd=sd(sus17))
```

On obtient un avertissement sur les «ties». Ceci est dû au fait que le test est pour des données continues et que les valeurs répétées obtenus dans le SUS sont un peu susceptibles.

La probabilité de faire une erreur en rejettant $H_0$ est très élevée, nous acceptons donc $H_0$.

Et la même chose pour 2018:

```r
ks.test(sus18, "pnorm", mean=mean(sus18), sd=sd(sus18))
```

avec le même résultat.

Faisons un test de Student pour comparer les deux volées. L'hypothèse à tester peut être formulée ainsi:

$H_0: \mu_1 - \mu_2 = 0$  
$H_1: \mu_1 - \mu_2 \neq 0$

```r
t.test(sus17, sus18)
```

La probabilité de commettre une erreur en rejettant $H_0$ est de plus de 50%. Nous ne pouvons donc pas rejeter $H_0$ et nous conculons qu'il n'y a pas de différence statistiquement significative entre les résultats du SUS de 2017 et 2018.
